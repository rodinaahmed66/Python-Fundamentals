```python
# Re-run this cell
# Preloaded data for validating email domain.
top_level_domains = [
    ".org",
    ".net",
    ".edu",
    ".ac",
    ".gov",
    ".com",
    ".io"
]
```

```python
# Start coding here. Use as many cells as you need.
def validate_name(name):
    if type(name)!=str:
        return False 
    elif len(name)<=2:
        return False 
    else:
        return True
```

```python
def validate_email(email):
    
    if '@' not in email:
       return False
    for i in top_level_domains:
        if i in email:
           return True 
     
    return False    
```
